## INFORME ANALISIS ESTADISTICA - GEPOL

1. Importar las librer√≠as o APIs necesarias para el procesamiento de datos.

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Leer los datos necesarios para graficarlos

```python
df = pd.read_json('C:/Users/wagne/Downloads/ANALISIS-ESTADISTICA-GEPOL/data/gepol.json')
df.head(100)
```

3. Ordenar los datos adecuadamente de una forma ascendente

```python
#organization by month o sea, organizacion por mes
by_month = df.sort_values('Mes')
by_month.head(10)
```

4. Graficar los datos con la librer√≠a Seaborn 

```python

sns.catplot(x = "Mes", data = df, kind = "count");
plt.title('CANTIDAD DE DOCUMENTOS PRODUCIDOS EN EL GEPOL MES 01 Y 02 - 2023') # T√≠tulo del gr√°fico.
```

5. Los datos graficados anteriormente, estan sin tanta informaci√≥n para su an√°lisis, por lo cual, se pondr√°n leyendas en las barras que muestran los datos analizados.

```python
groupedvalues = df.groupby('Mes').sum().reset_index()
g = sns.barplot(x = 'Mes', y = 'Cantidad', data = groupedvalues)
for index, row in groupedvalues.iterrows():
    g.text(row.name, row.Cantidad - 2, round(row.Cantidad, 2), color = 'red', ha = "center")

plt.title('CANTIDAD DE DOCUMENTOS PRODUCIDOS EN EL GEPOL MES 01 Y 02 - 2023')
```

6. Graficar un histograma 

```python
#histograma
sns.histplot(data=df, x="Cantidad", kde=True)
plt.title('CANTIDAD DE DOCUMENTOS PRODUCIDOS EN EL GEPOL MES 01 Y 02 - 2023')
```

7. Crear una gr√°fica de dispersi√≥n

```python
sns.stripplot(data=df, x="Cantidad", y="Mes")
plt.title('CANTIDAD DE DOCUMENTOS PRODUCIDOS EN EL GEPOL MES 01 Y 02 - 2023')
```

8. Crear una gr√°fica de viol√≠n

```python
sns.violinplot(data=df, x="Cantidad", y="Mes")
plt.title('CANTIDAD DE DOCUMENTOS PRODUCIDOS EN EL GEPOL MES 01 Y 02 - 2023')
```

7. gr√°fica tipo parcela (PairPlot)

```python
#graficar varios graficos o parcelas

sns.pairplot(df,height=1.5)

```

8. Gr√°fica tipo mixta (viol√≠n + dispersi√≥n)

```python
sns.catplot(data=df, x="Cantidad", y="Mes", kind="violin", color=".9", inner=None)
sns.swarmplot(data=df, x="Cantidad", y="Mes", size=3)
plt.title('CANTIDAD DE DOCUMENTOS PRODUCIDOS EN EL GEPOL MES 01 Y 02 - 2023')
```

9. Gr√°ficas de tipo dispersi√≥n, segmentando datos.

```python
sns.relplot(data=df, x="Cantidad", y="Mes", hue="Mes")
plt.title('CANTIDAD DE DOCUMENTOS PRODUCIDOS EN EL GEPOL MES 01 Y 02 - 2023')
```

10. Gr√°fica tipo linea, para exponer la cronolog√≠a de producci√≥n documental

```python
sns.relplot(data=df, kind="line")
plt.title('CANTIDAD DE DOCUMENTOS PRODUCIDOS EN EL GEPOL MES 01 Y 02 - 2023')
```
